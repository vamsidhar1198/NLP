```python
import pandas as pd 
import numpy as np
from gensim.models.word2vec import Word2Vec
import nltk
from nltk.tokenize import word_tokenize
```

```python
bbc_data = pd.read_csv(r"C:\Users\Home\Downloads\bbc-text.csv")
```

```python
bbc_data
```

There are 2225 datapoints and 2 cloumns 

 There are 5 unique category ('tech', 'business', 'sport', 'entertainment', 'politics'), for each category there are many text given 

```python
list_of_sent = []
for i in bbc_data['text']:
    list_of_sent.append(i.split())
```

i have split each and every word from the column text in bbc dataframe 

```python
w2v_model = Word2Vec(list_of_sent,vector_size=10,min_count=5)
```

here i am training the model with all words taken from text column in bbc_data dataframe as my input data size is less i have decreased the vectoer size of the words to 10 , if the word is occured minimun 5 times in total list_of_sent list then that word will be considered and trained to the model 

```python
words = list(w2v_model.wv.key_to_index)
```

this command will give all the words for which the model is trained and given vectors , i have stored all those words into a variable 'words'

```python
sent_vecs = []
for i in bbc_data['text'][0:250]:
    sent_array = []
    count = 0
    for j in word_tokenize(i):
        if j in words:
            sent_array.append(w2v_model.wv[j])
            count +=1
    sent_vecs.append(np.sum(sent_array,axis=0)/count)
```

in word2vec only the words are converted to vectors but we have convert whole sentance to vector , so we use one of the technique avgword2vec so that we can convert whole sentence in to array of vector

i took first 250 rows from the bbc_data and converted whole text column in those rows into vector form , here evey word dimension is 10 as we set in the above command so the dimension of sentance will be also 10 as we add up all the words in the sentance to get that sentance vector
